```python
def distribuicao(valor_max,p):
    res = {}

    q = valor_max // p  #quociente
    r = valor_max % p   #resto

    dif = 1/valor_max

    if q > 0:
        for i in range(p):
            res[i] = q * dif
    
    if r >= 0:
        i=0
        while r >= 0:
            try:
                res[i] += dif
            except:
                res[i] = dif
            i += 1
            r -= 1
    return res
    
res = distribuicao(2**64 -1, 251)
print(res)

#display(bar_chart([res[c] for c in range(0,256)]))
```
